```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
```

```python
tickers = ["FUBO", "NFLX", "AAPL", "GOOG", "UCO", "AMZN", "TSLA"]
stocks = yf.download(tickers, start="2013-09-08", end="2023-09-08")
table = stocks.loc[:,"Close"].copy()
table.head()
```

```python
table.plot(figsize= (20,10), fontsize=15)
pyplot.style.use("classic")
pyplot.show
```

```python
table = table.transpose()
```
