```python
Stocks = ["NFLX","TSLA","GOOG"]
Start = "2013-01-01"
End = "2023-01-01"

Data = yf.download(Stocks, start= Start, end= End)
```

```python

GOOG_dailies = Copy['GOOG'].pct_change().dropna()
GOOG_STD = GOOG_dailies.std()
GOOG_AV = GOOG_STD * np.sqrt(252)




NFLX_dailies = Copy['NFLX'].pct_change().dropna()
NFLX_STD = NFLX_dailies.std()
NFLX_AV = NFLX_STD * np.sqrt(252)




TSLA_dailies = Copy['TSLA'].pct_change().dropna()
TSLA_STD = TSLA_dailies.std()
TSLA_AV = TSLA_STD * np.sqrt(252)
print(f"Annualized Volatility for GOOG: {GOOG_AV:.2%}, for NFLX: {NFLX_AV:.2%},for TSLA: {TSLA_AV:.2%}")

```

```python
def calculate_annualized_volatility(stock_data):
    dailies = stock_data.pct_change().dropna()
    std_dev = dailies.std()
    annualized_volatility = std_dev * np.sqrt(252)
    return annualized_volatility

volatilities = {}
for stock in Stocks:
    volatilities[stock] = calculate_annualized_volatility(Copy[stock])

for stock, volatility in volatilities.items():
    print(f"Annualized Volatility for {stock}: {volatility:.2%}")
```
