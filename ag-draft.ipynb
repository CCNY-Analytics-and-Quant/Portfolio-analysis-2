```python
ticker = 'AAPL'

# downloading data
df = yf.download(ticker, period ="3mo")
df = df.reset_index()
```

```python
df.tail()
```

```python
# computing the logarithmic returns using the Closing price
df['Log-return'] = np.log(df['Close'] / df['Close'].shift(1))
df['Log-return'].fillna(0, inplace=True)
```

```python
df
```

```python
ROLLING_DAYS = 60
# compute the volatility using the pandas rolling standard deviation function
df['Annualized-Volatility'] = df['Log-return'].rolling(window=ROLLING_DAYS).std() * np.sqrt(ROLLING_DAYS)
```
